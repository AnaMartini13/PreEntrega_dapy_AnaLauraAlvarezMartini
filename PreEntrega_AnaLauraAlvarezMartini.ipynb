### 3锔 Calidad de los datos

```python
def calidad(df, nombre, pk=None): #PK Primary Key
  print(f'\033[1m*** {nombre} ***\033[0m\n')

  print(f'\033[1mValores nulos:\033[0m')
  display(df.isna().sum().to_frame('Nulos'))

  dup_rows = df.duplicated(keep=False).sum()
  print(f'\n\033[1mFilas duplicadas:\033[0m', dup_rows, '\n')

  if pk and pk in df.columns:
    dup_pk = df[pk].duplicated(keep=False).sum()
    print(f'\033[1mDuplicados por PK:\033[0m', dup_pk,'\n')
    if dup_pk > 0:
      duplicados = (
        df[df[pk].duplicated(keep=False)][pk]
        .value_counts()
        .sort_values(ascending=False)
        .reset_index()  #para que el atributo y el count aparezcan en la misma l铆nea
      )
      print(f'\033[1mTop 10 duplicados m谩s frecuentes:\033[0m')
      display(duplicados.head(10))
    else:
      print(f'No se encontraron duplicados en "{pk}".\n')
  else:
    if pk:
      print(f'La clave "{pk}" no existe en el DataFrame.')
    else:
      print('No se indic贸 una clave para analizar duplicados por columna.')



```

```python
#llamo a las funciones

calidad(ventas, 'VENTAS', pk='id_venta')
print('\n')
calidad(marketing, 'MARKETING', pk='id_campanha')
print('\n')
calidad(clientes, 'CLIENTES', pk='id_cliente')
```

### 4锔 Limpieza de datos

```python
#creo copia de dfs para no alterar el original
ventas_cp = ventas.copy() #cp de copia limpia
marketing_cp = marketing.copy()
clientes_cp = clientes.copy()
```

```python
#elimino duplicados y nulos (al faltar precio y cantidad en 2 registros)
ventas_cp = ventas_cp.drop_duplicates().dropna(subset=['precio', 'cantidad'])
marketing_cp = marketing_cp.drop_duplicates()
clientes_cp = clientes_cp.drop_duplicates()
```

```python
#valido
print(ventas_cp.dtypes)
print('\n')
print(marketing_cp.dtypes)
print('\n')
print(clientes_cp.dtypes)
```

```python
#TEXTOS

def normalizar_texto(df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.replace(r"[\u200b\t\r\n]", "", regex=True) #la r es para q lo interprete regex y no python
            .str.replace(" +", " ", regex=True)
            .str.title()
        )
    return df
```

```python
ventas_cp = normalizar_texto(ventas_cp)
marketing_cp = normalizar_texto(marketing_cp)
clientes_cp = normalizar_texto(clientes_cp)

display(ventas_cp.sample(5))
print('\n')
display(marketing_cp.sample(5))
print('\n')
display(clientes_cp.sample(5))
```

```python
#FECHAS

def normalizar_fechas(df):
    for col in df.columns:
        if "fecha" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
    return df
```

```python
ventas_cp["fecha_venta"] = pd.to_datetime(ventas_cp["fecha_venta"], errors="coerce", dayfirst=True)
marketing_cp["fecha_inicio"] = pd.to_datetime(marketing_cp["fecha_inicio"], errors="coerce", dayfirst=True)
marketing_cp["fecha_fin"] = pd.to_datetime(marketing_cp["fecha_fin"], errors="coerce", dayfirst=True)

print(ventas_cp['fecha_venta'].dtype)
print(marketing_cp['fecha_inicio'].dtype)
print(marketing_cp['fecha_fin'].dtype)
```

```python
#NUMRICOS

if "cantidad" in ventas_cp.columns:
    ventas_cp["cantidad"] = pd.to_numeric(ventas_cp["cantidad"], errors="coerce").astype("Int64")

if "precio" in ventas_cp.columns:
    ventas_cp["precio"] = (ventas_cp["precio"].astype(str).str.strip().str.replace("$", "", regex=False).str.replace(",", "", regex=False))
    ventas_cp["precio"] = pd.to_numeric(ventas_cp["precio"], errors="coerce") #coerce los convierte a NaN


print(ventas_cp['cantidad'].dtype)
print(ventas_cp['precio'].dtype)
```

```python
display(ventas_cp.sample(5))
print('\n')
display(marketing_cp.sample(5))
print('\n')
display(clientes_cp.sample(5))
```

```python
#REPORTE GLOBAL CALIDAD

def reporte_global_calidad(dfs, nombres):
    resumen = []
    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()
        duplicados = df.duplicated(keep=False).sum()
        columnas = len(df.columns)
        filas = len(df)
        resumen.append({"Dataset": nombre, "Filas": filas, "Columnas": columnas, "Total Nulos": nulos, "Duplicados": duplicados})
    reporte = pd.DataFrame(resumen)
    return reporte
```

```python
#valido
print(f'\033[1m*** REPORTE GLOBAL CALIDAD ***\033[0m\n')
display(reporte_global_calidad([ventas, marketing, clientes], ["VENTAS Original", "MARKETING Original", "CLIENTES Original"]))
print ('\n')
display(reporte_global_calidad([ventas_cp, marketing_cp, clientes_cp],["VENTAS limpio", "MARKETING limpio", "CLIENTES limpio"]))
```

### 5锔 Transformaci贸n de datos

```python
#busco columna de producto
def encontrar_col(df, patrones):
    for c in df.columns:
        nombre = c.lower()
        if any(p in nombre for p in patrones):
            return c
    return None

col_prod = encontrar_col(ventas_cp, ['producto', 'id_producto', 'sku', 'articulo', 'art铆culo'])
if col_prod is None:
    raise ValueError('No se encontr贸 columna de producto.')
print(col_prod)
```

```python
#INGRESO
perf_vtas = (ventas_cp.assign(ingreso = ventas_cp['precio'] * ventas_cp['cantidad']))
```

```python
#agrego
resumen_prod = (
    perf_vtas
    .groupby(by=col_prod, dropna=False, as_index=False, observed=False)
    .agg(y_total=('ingreso', 'sum'), unidades=('cantidad', 'sum'), precio_avg=('precio', 'mean'), registros=('ingreso', 'size'))
    .round({'precio_avg': 2})
    .sort_values(by='y_total', ascending=True)
)

display(resumen_prod.sample(10))
```

```python
#PERCENTIL 80
p80_y = resumen_prod['y_total'].quantile(q=0.80, interpolation='linear')
```

```python
#par: prod alto rend
par_vtas = resumen_prod.query('y_total >= @p80_y', engine='python').sort_values(by=['y_total', 'unidades'], ascending=[False, False], na_position='last', ignore_index=True)
```

```python
#valido
print(f'\033[1mColumna de producto detectada:\033[0m {col_prod}')
print(f'\033[1mPercentil 80 del Ingreso Total:\033[0m ${float(p80_y):,.2f}')
print('\n')
print('\033[1m Productos de ALTO RENDIMIENTO (top 20% por Ingreso):\033[0m')
display(par_vtas.head(20))
```

### 6锔 Agregaci贸n

```python
#busco columna categor铆a
col_cat = encontrar_col(ventas_cp, ['categoria', 'categor铆a', 'categoria_producto', 'rubro'])
if col_cat is None:
    raise ValueError('No se encontr贸 columna de categor铆a.')
print(col_cat)
```

```python
#INGRESO
if "ingreso" not in ventas_cp.columns:
    vtas_cat = ventas_cp.assign(ingreso = ventas_cp["precio"] * ventas_cp["cantidad"])
else:
    vtas_cat = ventas_cp.copy()
```

```python
#agrego
resumen_cat = (
    vtas_cat
    .groupby(by=col_cat, dropna=False, as_index=False)
    .agg(y_total=('ingreso', 'sum'), unidades=('cantidad', 'sum'), ventas=('ingreso', 'size'), precio_avg=('precio', 'mean'))
    .sort_values(by='y_total', ascending=False, na_position='last', ignore_index=True)
)
```

```python
#saco promedios
resumen_cat = (
    resumen_cat
    .assign(avg_ticket_por_vta = resumen_cat['y_total'] / resumen_cat['ventas'])
    .round({'avg_ticket_por_vta': 2, 'precio_avg': 2})
)
```

```python
#valido
print(f'\033[1mColumna de categor铆a detectada:\033[0m {col_cat}')
print('\n')
print('\033[1mResumen por categor铆a (seg煤n Ingreso Total):\033[0m')
display(resumen_cat.head(20))
```
